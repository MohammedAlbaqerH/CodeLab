```python
import pandas as pd
```

# Task 1
## Read CSV file DataFrame using pandas

```python
emplopyes = pd.read_csv('inputs/employees.csv')
emplopyes
```

# Task 2
## Drop first column

```python
emplopyes.drop('First Name', axis = 1, inplace=True)
emplopyes
```

# Task 3
## Fill missing values in (Salary, Bonus %) using the mean value

```python
emplopyes[['Salary', 'Bonus %']] = emplopyes[['Salary', 'Bonus %']].fillna(emplopyes.mean(axis = 0))
# mean(axis = 0) will give us the mean with recpact to columns
emplopyes
```

### Check if there is any NaN value

```python
emplopyes[emplopyes['Salary'].isna()]
```

```python
emplopyes[emplopyes['Bonus %'].isna()]
```

# Task 4
## Create a dictionary for the mapping of Team and Gender column

```python
def nameDicMaker(names):
    '''This function take a list of names and return dictionary
     such that the keys are names and, the values are number from 0 to length
     names list '''
    names_dic = {}
    num = 0
    for name in names:
        if name in names_dic: continue
        names_dic[name] = num
        num += 1
    return names_dic
```

```python
teams_dictionary = nameDicMaker(emplopyes['Team'])
print(teams_dictionary)
```

```python
gender_dictionary = nameDicMaker(emplopyes['Gender'])
print(gender_dictionary)
```

# Task 5
## Convert a text to numerical values in (Gender, Team)

```python
emplopyes['Team'] = emplopyes['Team'].apply(lambda name:teams_dictionary[name])
emplopyes['Gender'] = emplopyes['Gender'].apply(lambda name:gender_dictionary[name])
emplopyes
```
