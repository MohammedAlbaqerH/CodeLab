# Task 2
## split data into inputs/featurs (X) and output/target (y)

```
X = df.loc[:, (df.columns != 'diagnosis')&(df.columns != 'id')]
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

### Standard Scaling

```
normlizer = StandardScaler()
normlizer.fit(X)
X_train = normlizer.transform(X_train)
X_test = normlizer.transform(X_test)
```

```
model = LogisticRegression(C = 0.3, random_state = 42)
```

# Task 3, 4
Train model and compute accuracy

```
model.fit(X_train, y_train)
```

```

print(f'accuracy = {round(model.score(X_test, y_test),4)*100}%')
```
